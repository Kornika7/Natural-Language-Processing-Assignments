```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
```

Data cleaning

```python
import re
import string

def clean_text(text ):
    text =''.join(word for word in text if word not in string.punctuation)
    text1 = ''.join(w.lower() for w in text if not w.isdigit())
    return text1
```

Tokenization and Lemmatization

```python
# Lemmatization of text data
def lemmatize_text(text):
    wordlist = []
    lemmatizer = WordNetLemmatizer()
    # Sentence tokenization
    sentences = nltk.sent_tokenize(str(text))

    intial_sentences = sentences[0:1]
    final_sentences = sentences[len(sentences)-2: len(sentences)-1]

    # Word tokenization
    for sentence in intial_sentences:
        words = nltk.word_tokenize(str(sentence))
        for word in words:
            wordlist.append(word)

    for sentence in final_sentences:
        words = nltk.word_tokenize(str(sentence))
        for word in words:
            wordlist.append(word)

    return wordlist
```

Applying Functions on train data

```python
# Apply the functions on title column

# Clean
df['title'] = df['title'].apply(lambda x: clean_text(x))
# Lemmatize
df['title'] = df['title'].apply(lambda x: lemmatize_text(x))
```

```python
# Apply the functions on text column

# Clean
df['text'] = df['text'].apply(lambda x: clean_text(x))
# Lemmatize
df['text'] = df['text'].apply(lambda x: lemmatize_text(x))
```

```python
# Load GoogleNews Word2Vec (downloaded from https://code.google.com/archive/p/word2vec/)
from gensim.models import KeyedVectors
word2vec_model_path = '/path/to/GoogleNews-vectors-negative300.bin.gz'
word2vec_model = KeyedVectors.load_word2vec_format(word2vec_model_path, binary=True)

# Function to convert text to vectors using Word2Vec
def text_to_vectors(text):
    words = nltk.word_tokenize(text)
    vectors = []
    for word in words:
        if word in word2vec_model:
            vectors.append(word2vec_model[word])
    return vectors

# Apply the function to the 'text' column
df['text_vectors'] = df['text'].apply(text_to_vectors)
```
